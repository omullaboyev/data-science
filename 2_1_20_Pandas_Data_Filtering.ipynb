```python
df["model"].unique() # models ga df ning model ustunidagi takrorlanmas qiymatlar o'zlashtirib olamiz
```

```python
models = [" SLK", " B Class"] # bizga kerakli modellar ro'yxati
filtr = df["model"].isin(models) # df ning model ustunidagi qiymat bizga kerakli modellar ro'yxatida bo'lsa
filtr
```

```python
df[filtr] # filtrimizni df ga qo'llaymiz
```

```python
# yoki filtrlangan ma'lumotlarni yangi df2 ga saqlab olishimiz mumkin
df2 = df[filtr]
df2
```

```python
filtr = df["fuelType"].isin(["Diesel", "Petrol"]) # fuelType ustuni qiymati faqat Diesel va Petrol bo'lgan elementlar uchun filtr yaratamiz
df[filtr] # filtrimizni df ga qo'llaymiz
```

```python
 df[df.year > 2018] # year ustunidagi qiymat 2018 dan katta bo'lgan elementlarni filtrlab olish
```

```python
df[df.year >= 2018] # year ustunidagi qiymat 2018 dan katta yoki teng bo'lgan elementlarni filtrlab olish
```

```python
df[(df.year >= 2018) & (df.price < 20000)] # year ustunidagi qiymat 2018 dan katta yoki teng va price ustunidagi qiymat 20000 dan kichik bo'lgan elementlarni filtrlab olish
```

```python
df3 = df[(df.year >= 2018) & (df.price < 20000) & (df.mileage < 30000)] # year >= 2018, price < 20000, mileage < 30000 bo'lgan elementlarni df3 ga saqlab olamiz
df3
```

```python
df[df.transmission == "Manual"] # transmission ustunidagi qiymat Manual bo'lgan qatorlar
```
