2.1.16 Reytinglash (Ranking)

**1 - eng baland reyting**

```python
import pandas as pd
import numpy as np
```

```python
# internetdan CSV faylni dataframe ga yuklab olamiz
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv", index_col = 0)
df.head()
```

```python
# mashinalarga ishlab chiqarilgan yili bo'yicha reyting beramiz
df["rating"] = df["year"].rank()
```

```python
df
```

Har bir mashinaning **year** bo'yicha reytingi df da shu yilda ishlab chiqarilgan mashina nechta bo'lsa shunga teng.

```python
df.sort_values("year") # year ustuni qiymatlari bo'yicha tartiblaymiz
```

*   1973 yilda ishlab chiqarilgan mashina bitta bo'lgani uchun reytingi 1 ga teng (eng baland reyting)
*   1984 yilda ishlab chiqarilgan mashina ikkita bo'lgani uchun reytingi 2 ga teng
*   1993 yilda ishlab chiqarilgan mashina uchta bo'lgani uchun reytingi 3 ga teng
*   .............
*   2020 yilda ishlab chiqarilgan mashina ko'p bo'lgani uchun reytingi 2475.5 ga teng (eng past reyting)

#`method` - reytinglash metodi
min - eng kichikdan
average - o'rtacha qiymat

```python
# mashinalarga narxi bo'yicha reyting beramiz
df["rating"] = df["price"].rank(method="min") # minimumdan boshlab reyting beramiz
```

```python
df
```

```python
df.sort_values("price") # price - narxi bo'yicha tartiblaymiz.
```

```python
# mashinalarga mileage bo'yicha reyting beramiz
df["rating"] = df["mileage"].rank(method="average") # o'rtacha qiymat bo'yicha reyting beramiz
```

```python
df.sort_values("mileage") # mileage bo'yicha tartiblaymiz.
```
