2.1.15 Tartiblash (Sort)

```python
import pandas as pd
import numpy as np
```

#Series obyektini tartiblash

```python
obj = pd.Series(range(4), index = ["d", "a", "b", "c"])
obj
```

#`sort_index` - Series ni indeks bo'yicha tartiblash

```python
obj.sort_index()
```

#`sort_values` - Series ni qiymatlar bo'yicha tartiblash

```python
obj.sort_values()
```

#DataFrame obyektini tartiblash

```python
# internetdan CSV faylni dataframe ga yuklab olamiz
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv", index_col = 0)
df.head()
```

```python
df.shape
```

#`sort_index` - DataFrame ni index bo'yicha tartiblash

```python
df.sort_index() # indeksi bo'yicha tartiblash
```

```python
df.sort_index(axis = 1) # ustunlar bo'yicha tartiblash
```

```python
df.sort_index(axis = 1, inplace = True) # ustunlar bo'yicha tartiblash, inplace = True DataFrame tartiblangan holda qoladi
```

```python
df.sort_index(ascending = False) # Indeks bo'yicha teskari tartiblash
```

```python
df.sort_index(axis = 1, ascending = False) # Ustunlar bo'yicha teskari tartiblash
```

#`sort_values` - DataFrame ni qiymatlar bo'yicha tartiblash

```python
df.sort_values(by = "price") # berilgan ustunnning qiymatlari bo'yicha tartiblash
```

```python
df.sort_values(by = "price", ascending = False) # berilgan ustunnning qiymatlari bo'yicha teskari tartiblash
```

```python
df.sort_values(by = ["price", "mileage"], ascending = False) # price va milage ustunlarining qiymatlari bo'yicha teskari tartiblash
```

```python
df.sort_values(by = "year") # year ustunidagi qiymatlar bo'yicha tartiblash
```
