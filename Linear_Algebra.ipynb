#Matritsalarni ko'paytirish

#Shart 1: Birinchi (ko'payuvchi) matritsaning ustunlari soni ikkinchi (ko'paytiruvchi)matritsaning qatorlari soniga teng bo'lishi kerak

image.png

Natija matritsaning o'lchami:


*   Qatorlar soni (axis 0): Ko'payubchi matritsaning qatorlarin soni
*   Ustunlar soni (axis 1): Ko'paytiruvchi matritsaning ustunlari soni




```python
# 1-matritsa
x = np.array([[1., 2., 3.], [4., 5., 6]])
x
```

```python
# 2-matritsa
y = np.array([[6., 23.], [-1., 7.], [8., 9.]])
y
```

```python
x.dot(y) # x*y. x - ko'payuvchi, y - ko'paytiruvchi
#x[0,0]*y[0,0] + x[0,1]*y[0,1]+x[0,2]*y[0,2]  x[0,0]*y[1,0] + x[0,1]*y[1,1]+x[0,2]*y[1,2]
#x[1,0]*y[0,0] + x[1,1]*y[0,1]+x[1,2]*y[0,2]  x[1,0]*y[1,0] + x[1,1]*y[1,1]+x[1,2]*y[1,2]
#=>
#[6-2+24  23+14+27
# 24-5+48  92+35+54]
#=>
#[28 64
# 67 181]
```

```python
np.dot(x, y) # x*y
```

```python
y.dot(x) # y*x. y - ko'payuvchi, x - ko'paytiruvchi
```

```python
np.dot(y, x) # y*x. y - ko'payuvchi, x - ko'paytiruvchi
```
